```python
def show_correlation(df, size=(15,10)):
    corr = df.corr()
    fig, ax = plt.subplots()
    fig.set_size_inches(size[0], size[1])
    sns.heatmap(corr, annot=True, cmap='PiYG')
```

```python
test = pd.read_csv("data/test.csv")
train = pd.read_csv("data/train.csv")
```

```python
train.drop(['Survived'], axis=1).describe()
```

```python
test.describe()
```

```python
perc = train.Age.isnull().sum()*100/train.Age.count()
print('{}% of Ages is null'.format(perc))
```

```python
train.Age.count()
```

```python
le = LabelEncoder()
le.fit(train.Sex)

train.Sex = le.transform(train.Sex)
test.Sex = le.transform(test.Sex)
train.head()
```

```python
train = train.fillna({'Cabin': 0})
test = test.fillna({'Cabin': 0})

train.loc[train.Cabin != 0, 'Cabin'] = 1
test.loc[train.Cabin != 0, 'Cabin'] = 0

train.head()
```

```python
train.Embarked.isnull().sum()
```

```python
embarked_encoder = LabelEncoder()
common_embarked = train.groupby(['Embarked'])['Embarked'].value_counts().idxmax()[0]
train = train.fillna({'Embarked': common_embarked})
test = test.fillna({'Embarked': common_embarked})
embarked_encoder.fit(train['Embarked'].unique())
train['Embarked'] = embarked_encoder.transform(train['Embarked'])
test['Embarked'] = embarked_encoder.transform(test['Embarked'])
```
